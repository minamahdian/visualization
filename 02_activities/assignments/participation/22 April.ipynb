```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize= (5,3))
ax.hist (y)
```

```python
fig, ax = plt. subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growht over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family': 'monospace','color':'green', 'size':14}
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Totoal growth over time', fontdict = font1, loc='left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
#modifying our datapoints
fig, ax = plt. subplots(figsize = (5,3))

ax.scatter(x,y,
           marker = '*',
           color='indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y, marker='*',
        color='indigo',
        linestyle = '--',
        linewidth = 2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y,
        marker= '*',
        markersize = 12,
        color ='#7425b9',
        linestyle='--',
        linewidth=2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000'
        )
ax.grid(axis = 'y', color='blue', linewidth = 2, linestyle='-.')
```
